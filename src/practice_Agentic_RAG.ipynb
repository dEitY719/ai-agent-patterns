### 3. ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œ ì •ì˜ (Handler Functions)

ì´ ì½”ë“œëŠ” ì§ˆë¬¸ ìœ í˜•ë³„ë¡œ **íŠ¹í™”ëœ ì²˜ë¦¬ë¥¼ ìˆ˜í–‰í•˜ëŠ” ë…¸ë“œ í•¨ìˆ˜**ë¥¼ ì •ì˜í•©ë‹ˆë‹¤. ê° í•¨ìˆ˜ëŠ” `SimpleState` ìƒíƒœë¥¼ ì…ë ¥ë°›ì•„, ë‹µë³€ì„ ìƒì„±í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤.
- handle_translation: ë²ˆì—­ ê´€ë ¨ ì§ˆë¬¸ ì²˜ë¦¬
- handle_math: ìˆ˜í•™/ê³„ì‚° ê´€ë ¨ ì§ˆë¬¸ ì²˜ë¦¬
- handle_general: ì¼ë°˜ Q&A ì§ˆë¬¸ ì²˜ë¦¬
> ê° ë…¸ë“œëŠ” ì…ë ¥ëœ ìƒíƒœë¥¼ ê¸°ë°˜ìœ¼ë¡œ ì ì ˆí•œ ë‹µë³€ì„ ìƒì„±í•˜ë©°, ì´í›„ ì¡°ê±´ë¶€ ë¼ìš°íŒ… ì‹œìŠ¤í…œì—ì„œ ì„ íƒì ìœ¼ë¡œ í˜¸ì¶œë©ë‹ˆë‹¤.

````python
def handle_translation(state: SimpleState):
    return {"answer": f"ë²ˆì—­ ì „ë¬¸ ì‹œìŠ¤í…œì´ '{state['question']}'ì„ ì²˜ë¦¬í–ˆìŠµë‹ˆë‹¤."}


def handle_math(state: SimpleState):
    return {"answer": f"ìˆ˜í•™ ê³„ì‚° ì‹œìŠ¤í…œì´ '{state['question']}'ì„ ì²˜ë¦¬í–ˆìŠµë‹ˆë‹¤."}


def handle_general(state: SimpleState):
    return {"answer": f"ì¼ë°˜ ì§ˆì˜ì‘ë‹µ ì‹œìŠ¤í…œì´ '{state['question']}'ì„ ì²˜ë¦¬í–ˆìŠµë‹ˆë‹¤."}
````

### 4. ì§ˆë¬¸ ë¶„ë¥˜ ë…¸ë“œ (Router / Classifier)

ì´ ë…¸ë“œëŠ” ì‚¬ìš©ìì˜ ì§ˆë¬¸ì„ ë³´ê³  **ì§ˆë¬¸ ìœ í˜•ì„ íŒë‹¨**í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤.

- ì§ˆë¬¸ì— `"ë²ˆì—­"` ë˜ëŠ” `"translate"`ê°€ í¬í•¨ë˜ë©´ â†’ `translation`  
- ì§ˆë¬¸ì— `"ê³„ì‚°"` ë˜ëŠ” `"ìˆ˜í•™"`ì´ í¬í•¨ë˜ë©´ â†’ `math`  
- ê·¸ ì™¸ì˜ ì§ˆë¬¸ì€ â†’ `general`  

ë°˜í™˜ê°’ì€ ë‹¤ìŒ ë‹¨ê³„ì—ì„œ ì‚¬ìš©í•  `question_type` ì •ë³´ë¥¼ ë‹´ê³  ìˆìœ¼ë©°, ì´ë¥¼ ê¸°ë°˜ìœ¼ë¡œ AgentëŠ” ì ì ˆí•œ ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œë¥¼ ì„ íƒí•©ë‹ˆë‹¤.

> ì¦‰, ì‚¬ìš©ìì˜ ì§ˆë¬¸ì„ ì½ê³  "ì–´ë–¤ ì²˜ë¦¬ ì‹œìŠ¤í…œì´ ì²˜ë¦¬í•´ì•¼ í• ì§€"ë¥¼ ê²°ì •í•˜ëŠ” **ë¶„ê¸°ì ** ì—­í• ì„ í•©ë‹ˆë‹¤.

````python
# ì§ˆë¬¸ ë¶„ë¥˜ ë…¸ë“œ


def classify_question(state: SimpleState):
    question = state["question"].lower()

    if "ë²ˆì—­" in question or "translate" in question:
        question_type = "translation"
    elif "ê³„ì‚°" in question or "ìˆ˜í•™" in question:
        question_type = "math"
    else:
        question_type = "general"

    return {"question_type": question_type}
````

### 5. ë¼ìš°í„° í•¨ìˆ˜ (Routing by Question Type)

ì´ í•¨ìˆ˜ëŠ” ì´ë¯¸ ë¶„ë¥˜ëœ `question_type`ì„ ê¸°ë°˜ìœ¼ë¡œ **ì§ˆë¬¸ì„ ì–´ëŠ ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œë¡œ ë³´ë‚¼ì§€ ê²°ì •**í•©ë‹ˆë‹¤.

- `translation` â†’ ë²ˆì—­ ì²˜ë¦¬ ë…¸ë“œ  
- `math` â†’ ìˆ˜í•™ ê³„ì‚° ì²˜ë¦¬ ë…¸ë“œ  
- `general` â†’ ì¼ë°˜ Q&A ì²˜ë¦¬ ë…¸ë“œ  

> ì¦‰, ì§ˆë¬¸ ìœ í˜•ì— ë”°ë¼ ì›Œí¬í”Œë¡œìš°ì˜ **ë‹¤ìŒ ì‹¤í–‰ ë…¸ë“œë¥¼ ì„ íƒ**í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤.

````python
# ë¼ìš°í„° í•¨ìˆ˜ - ì§ˆë¬¸ ìœ í˜•ì— ë”°ë¼ ë¶„ê¸°


def route_question(state: SimpleState):
    question_type = state["question_type"]

    if question_type == "translation":
        return "translation"
    elif question_type == "math":
        return "math"
    else:
        return "general"
````

### 6. ê·¸ë˜í”„ êµ¬ì„± (StateGraph ìƒì„± ë° ë…¸ë“œ ì¶”ê°€)

ì´ ë‹¨ê³„ì—ì„œëŠ” **ì§ˆë¬¸ ì²˜ë¦¬ ì›Œí¬í”Œë¡œìš°ë¥¼ ê·¸ë˜í”„ í˜•íƒœë¡œ êµ¬ì„±**í•©ë‹ˆë‹¤.

1. `StateGraph(SimpleState)`ë¥¼ í†µí•´ ìƒíƒœ ê¸°ë°˜ ê·¸ë˜í”„ ìƒì„±  
2. ì£¼ìš” ë…¸ë“œ ì¶”ê°€:
   - `classifier`: ì§ˆë¬¸ ìœ í˜• ë¶„ë¥˜ ë…¸ë“œ  
   - `translation_handler`: ë²ˆì—­ ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œ  
   - `math_handler`: ìˆ˜í•™ ê³„ì‚° ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œ  
   - `general_handler`: ì¼ë°˜ ì§ˆì˜ì‘ë‹µ ë…¸ë“œ  

> ì´ë ‡ê²Œ êµ¬ì„±ëœ ê·¸ë˜í”„ëŠ” ì‚¬ìš©ìì˜ ì§ˆë¬¸ íë¦„ì— ë”°ë¼ ê° ë…¸ë“œë¥¼ ìˆœì°¨ì  í˜¹ì€ ì¡°ê±´ë¶€ë¡œ í˜¸ì¶œí•  ìˆ˜ ìˆëŠ” ê¸°ë°˜ì´ ë©ë‹ˆë‹¤.

````python
# ê·¸ë˜í”„ êµ¬ì„±
workflow = StateGraph(SimpleState)

# ë…¸ë“œ ì¶”ê°€
workflow.add_node("classifier", classify_question)
workflow.add_node("translation_handler", handle_translation)
workflow.add_node("math_handler", handle_math)
workflow.add_node("general_handler", handle_general)
````

### 7. ì—£ì§€ ì—°ê²° ë° ì¡°ê±´ë¶€ ë¶„ê¸°

ì´ ë‹¨ê³„ì—ì„œëŠ” **ê·¸ë˜í”„ ë…¸ë“œ ê°„ íë¦„**ê³¼ **ì¡°ê±´ë¶€ ë¶„ê¸°**ë¥¼ ì„¤ì •í•©ë‹ˆë‹¤.

   - `classifier` ë…¸ë“œì˜ ê²°ê³¼(`question_type`)ë¥¼ `route_question` ë¼ìš°í„° í•¨ìˆ˜ë¡œ ë¶„ì„  
   - ë°˜í™˜ê°’ì— ë”°ë¼ ë‹¤ìŒ ë…¸ë“œë¥¼ ì„ íƒ:
     - `translation` â†’ `translation_handler`  
     - `math` â†’ `math_handler`  
     - `general` â†’ `general_handler`  

3. **ëª¨ë“  ì²˜ë¦¬ ë…¸ë“œ ì¢…ë£Œ ì—°ê²°**
   - ê° ì²˜ë¦¬ ë…¸ë“œê°€ ì™„ë£Œë˜ë©´ `END` ë…¸ë“œë¡œ ì´ë™  

> ì´ë ‡ê²Œ ì„¤ì •í•˜ë©´, ì‚¬ìš©ìì˜ ì§ˆë¬¸ ìœ í˜•ì— ë”°ë¼ ì›Œí¬í”Œë¡œìš°ê°€ **ìë™ìœ¼ë¡œ ì ì ˆí•œ ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œë¡œ ë¶„ê¸°**ë˜ê³ , ìµœì¢…ì ìœ¼ë¡œ ì¢…ë£Œë©ë‹ˆë‹¤.

````python
# ì—£ì§€ ì—°ê²°
workflow.add_edge(START, "classifier")

# ì¡°ê±´ë¶€ ì—£ì§€ ì¶”ê°€ - í•µì‹¬ ë¶€ë¶„!
workflow.add_conditional_edges(
    "classifier",  # ë¶„ë¥˜ê¸°ì—ì„œ
    route_question,  # ë¼ìš°í„° í•¨ìˆ˜ë¡œ ë‹¤ìŒ ë…¸ë“œ ê²°ì •
    {  # ë°˜í™˜ê°’ì„ ì‹¤ì œ ë…¸ë“œì— ë§¤í•‘
        "translation": "translation_handler",
        "math": "math_handler",
        "general": "general_handler",
    },
)

# ëª¨ë“  ì²˜ë¦¬ ë…¸ë“œì—ì„œ ENDë¡œ ì—°ê²°
workflow.add_edge("translation_handler", END)
workflow.add_edge("math_handler", END)
workflow.add_edge("general_handler", END)
````

   - LLMì´ ì§ˆë¬¸ì„ ì½ê³  ì ì ˆí•œ í•¸ë“¤ëŸ¬ë¥¼ ì„ íƒí•˜ëŠ” ë‘ë‡Œ ì—­í•  ìˆ˜í–‰  
   - `temperature=0`ìœ¼ë¡œ ì„¤ì •í•˜ë©´ ê²°ì •ì´ ì¼ê´€ì ì´ë©°, ë†’ì´ë©´ ëœë¤ì„±ì´ ì¦ê°€

3. **Output Parser**
   - `JsonOutputParser`ë¥¼ ì‚¬ìš©í•´ LLMì˜ ì¶œë ¥(JSON) â†’ Python ê°ì²´ ë³€í™˜

4. **Router ì²´ì¸**
   - `router_chain = router_llm | classification_parser`  
   - LLM ì¶œë ¥ê³¼ íŒŒì‹±ì„ í•˜ë‚˜ì˜ ì²´ì¸ìœ¼ë¡œ ì—°ê²°

5. **Router ë¶„ë¥˜ í•¨ìˆ˜ (`router_classify`)**
   - ì…ë ¥ ì§ˆë¬¸ì„ ë°›ì•„ LLMì—ê²Œ ì „ë‹¬  
   - JSON í˜•íƒœë¡œ `question_type`ê³¼ `reasoning` ì¶”ì¶œ  
   - ì˜¤ë¥˜ ë°œìƒ ì‹œ ê¸°ë³¸ê°’ìœ¼ë¡œ `general` ì„ íƒ

6. **Router ê¸°ë°˜ ë‹¤ìŒ ë…¸ë“œ ê²°ì •**
   - `router_decide_next` í•¨ìˆ˜ê°€ ë¶„ë¥˜ ê²°ê³¼(`question_type`)ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ë‹¤ìŒ ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œ ì„ íƒ

7. **ì›Œí¬í”Œë¡œìš° êµ¬ì„± (StateGraph)**
   - `router` ë…¸ë“œ â†’ ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œ(`translation_handler`, `math_handler`, `general_handler`) â†’ `END`  
   - `add_conditional_edges`ë¡œ Router LLM ê²°ê³¼ì— ë”°ë¼ ì¡°ê±´ë¶€ ë¶„ê¸° ì—°ê²°

8. **ê·¸ë˜í”„ ì»´íŒŒì¼**
   - `improved_app = improved_workflow.compile()`  
   - ì‹¤í–‰ ê°€ëŠ¥í•œ ìƒíƒœ ê¸°ë°˜ ì›Œí¬í”Œë¡œìš° ìƒì„±


#### ì „ì²´ íë¦„ ìš”ì•½

1. ì‚¬ìš©ìê°€ ì§ˆë¬¸ ì…ë ¥
2. `router` ë…¸ë“œì—ì„œ LLMì´ ì§ˆë¬¸ ë¶„ì„ ë° ë¶„ë¥˜  
3. `router_decide_next` í•¨ìˆ˜ê°€ ì ì ˆí•œ ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œ ê²°ì •  
4. í•´ë‹¹ ì „ë¬¸ ì²˜ë¦¬ ë…¸ë“œ(`translation_handler`, `math_handler`, `general_handler`) ì‹¤í–‰  
5. ê²°ê³¼ë¥¼ `END` ë…¸ë“œë¡œ ì „ë‹¬í•˜ë©° ì›Œí¬í”Œë¡œìš° ì¢…ë£Œ  

> ê¸°ì¡´ í‚¤ì›Œë“œ ê¸°ë°˜ ë¼ìš°íŒ…ë³´ë‹¤ **ë” ìì—°ìŠ¤ëŸ¬ìš´ ì§ˆë¬¸ ì´í•´ì™€ ë¶„ë¥˜**ê°€ ê°€ëŠ¥í•˜ë©°, LLMì´ ë¶„ë¥˜ ì´ìœ (reasoning)ê¹Œì§€ ì œê³µí•˜ë¯€ë¡œ ë””ë²„ê¹… ë° ë¶„ì„ì—ë„ ìœ ë¦¬í•©ë‹ˆë‹¤.

````python
# Router LLMì„ ì‚¬ìš©í•œ ë¶„ë¥˜ ì‹œìŠ¤í…œ
from typing import Any, Dict

from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import JsonOutputParser
from langchain_openai import ChatOpenAI
from langgraph.graph import END, START, StateGraph
from pydantic import BaseModel, Field


# Pydantic ëª¨ë¸ ì •ì˜
class QuestionClassification(BaseModel):
    question_type: str = Field(description="ì§ˆë¬¸ ìœ í˜• (translation, math, general)")
    reasoning: str = Field(description="ë¶„ë¥˜ ì´ìœ ")


# Router LLM ìƒì„±
def create_router_llm():
    """ì§ˆë¬¸ì„ ë¶„ì„í•˜ê³  ì ì ˆí•œ í•¸ë“¤ëŸ¬ë¥¼ ì„ íƒí•˜ëŠ” Router LLM"""
    return ChatOpenAI(
        model="gpt-4o-mini",
        temperature=0,
    )


router_llm = create_router_llm()
# Output Parser ì„¤ì •
classification_parser = JsonOutputParser(pydantic_object=QuestionClassification)


# Router ì²´ì¸ ìƒì„±
router_chain = router_llm | classification_parser


# Router ë¶„ë¥˜ í•¨ìˆ˜
def router_classify(state: SimpleState) -> Dict[str, Any]:
    """Router LLMì´ ì§ˆë¬¸ì„ ë¶„ì„í•˜ê³  ì ì ˆí•œ í•¸ë“¤ëŸ¬ë¥¼ ê²°ì •"""

    question = state.get("question", "")

    # JSON ì¶œë ¥ì„ ìœ„í•œ í”„ë¡¬í”„íŠ¸
    classification_prompt = f"""ë‹¤ìŒ ì§ˆë¬¸ì„ ë¶„ì„í•˜ì—¬ ê°€ì¥ ì ì ˆí•œ í•¸ë“¤ëŸ¬ë¥¼ ì„ íƒí•˜ì„¸ìš”:

ì§ˆë¬¸: {question}

ë¶„ë¥˜ ê¸°ì¤€:
- ë²ˆì—­ ê´€ë ¨ ì§ˆë¬¸ â†’ translation
- ìˆ˜í•™ ê³„ì‚° ê´€ë ¨ ì§ˆë¬¸ â†’ math  
- ê¸°íƒ€ ì¼ë°˜ì ì¸ ì§ˆë¬¸ â†’ general

ë°˜ë“œì‹œ JSON í˜•ì‹ìœ¼ë¡œ ì‘ë‹µí•˜ì„¸ìš”:
{{
    "question_type": "ì„ íƒëœ_í•¸ë“¤ëŸ¬_íƒ€ì…",
    "reasoning": "ë¶„ë¥˜ ì´ìœ  ì„¤ëª…"
}}"""

    try:
        parsed_output = router_chain.invoke([HumanMessage(content=classification_prompt)])
        question_type = parsed_output["question_type"]
        reasoning = parsed_output["reasoning"]

    except Exception as e:
        print(f"Router ì˜¤ë¥˜: {e}")
        question_type = "general"  # ì˜¤ë¥˜ ì‹œ ê¸°ë³¸ê°’
        reasoning = "íŒŒì‹± ì˜¤ë¥˜ë¡œ ì¸í•œ ê¸°ë³¸ ì„ íƒ"

    return {
        "question_type": question_type,
        "reasoning": reasoning,
    }


# Router ë‹¤ìŒ ë…¸ë“œ ê²°ì • í•¨ìˆ˜
def router_decide_next(state: SimpleState) -> str:
    """Router ë¶„ë¥˜ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ë‹¤ìŒ ë…¸ë“œ ê²°ì •"""
    return state["question_type"]


# ê°œì„ ëœ ì›Œí¬í”Œë¡œìš° ìƒì„±
improved_workflow = StateGraph(SimpleState)

# ë…¸ë“œ ì¶”ê°€
improved_workflow.add_node("router", router_classify)
improved_workflow.add_node("translation_handler", handle_translation)
improved_workflow.add_node("math_handler", handle_math)
improved_workflow.add_node("general_handler", handle_general)

# ì—£ì§€ ì—°ê²°
improved_workflow.add_edge(START, "router")

# Router LLM ê¸°ë°˜ ì¡°ê±´ë¶€ ì—£ì§€ ì¶”ê°€
improved_workflow.add_conditional_edges(
    "router",  # Routerì—ì„œ
    router_decide_next,  # Router ê²°ì • í•¨ìˆ˜
    {  # ë°˜í™˜ê°’ì„ ì‹¤ì œ ë…¸ë“œì— ë§¤í•‘
        "translation": "translation_handler",
        "math": "math_handler",
        "general": "general_handler",
    },
)

# ëª¨ë“  ì²˜ë¦¬ ë…¸ë“œì—ì„œ ENDë¡œ ì—°ê²°
improved_workflow.add_edge("translation_handler", END)
improved_workflow.add_edge("math_handler", END)
improved_workflow.add_edge("general_handler", END)

# ê°œì„ ëœ ê·¸ë˜í”„ ì»´íŒŒì¼
improved_app = improved_workflow.compile()
````

ê°ê°ì˜ AgentëŠ” íŠ¹ì • ë„êµ¬ì— íŠ¹í™”ë˜ì–´ ë³‘ë ¬ë¡œ ì‘ì—…ì„ ìˆ˜í–‰í•˜ë©°, Supervisor Agentê°€ ì „ì²´ì ì¸ ì›Œí¬í”Œë¡œìš°ë¥¼ ì¡°ìœ¨í•©ë‹ˆë‹¤.

### 1. ì‹œìŠ¤í…œ êµ¬ì¡°:
- **Supervisor Agent**: ì‚¬ìš©ì ì§ˆë¬¸ì„ ë¶„ì„í•˜ê³  ì ì ˆí•œ ì „ë¬¸ Agentì—ê²Œ ì‘ì—…ì„ ë¶„ë°°

````python
import operator
from typing import Annotated, List, TypedDict

from langchain_core.messages import BaseMessage
from langchain_openai import ChatOpenAI


# State ì •ì˜ - ê° Agent ê°„ ì •ë³´ ê³µìœ ë¥¼ ìœ„í•œ ìƒíƒœ
class AgentState(TypedDict):
    messages: Annotated[List[BaseMessage], operator.add]
    next: str
    user_input: str
    selection_reasoning: str
    ml_result: str
    web_result: str
    url_result: str
    translation_result: str
    final_answer: str


# LLM ì •ì˜
llm_for_tools = ChatOpenAI(model="gpt-4o-mini", temperature=0)
````

### 2. Tool ì •ì˜

````python
from langchain.chains.summarize import load_summarize_chain
from langchain.tools import tool
from langchain.tools.retriever import create_retriever_tool
from langchain_community.document_loaders import WebBaseLoader
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_tavily import TavilySearch

# --- ë„êµ¬ 1: RAG ê²€ìƒ‰ ë„êµ¬ ---
retriever_tool = create_retriever_tool(
    retriever,
    "DeepLearningSearch",
    "ë¨¸ì‹ ëŸ¬ë‹ì— ê´€ë ¨ëœ ì§ˆë¬¸ì„ ê²€ìƒ‰í•˜ê³  ë°˜í™˜í•©ë‹ˆë‹¤. ë¨¸ì‹ ëŸ¬ë‹ì´ë‚˜ ë”¥ëŸ¬ë‹ì— ê´€ë ¨ëœ ì§ˆë¬¸ì— ì‚¬ìš©í•´ì•¼ í•©ë‹ˆë‹¤. ìµœì‹  ë”¥ëŸ¬ë‹ ë…¼ë¬¸ì— ëŒ€í•œ ë‚´ìš©ì€ í¬í•¨ë˜ì§€ ì•Šì„ ìˆ˜ë„ ìˆìŠµë‹ˆë‹¤.",
)

# --- ë„êµ¬ 2: ì›¹ ê²€ìƒ‰ ë„êµ¬ ---
tavily_tool = TavilySearch(
    k=5,
    description="ìµœì‹  ì •ë³´ë‚˜ ì¼ë°˜ì ì¸ ì‚¬ì‹¤ì— ëŒ€í•œ ì§ˆë¬¸ì— ë‹µí•˜ê¸° ìœ„í•´ ì›¹ì„ ê²€ìƒ‰í•©ë‹ˆë‹¤. ë²¡í„° ìŠ¤í† ì–´ì— ì—†ëŠ” ë‚´ìš©ì— ì‚¬ìš©ë©ë‹ˆë‹¤.",
)


# --- ë„êµ¬ 3: ì›¹í˜ì´ì§€ ë¶„ì„/ìš”ì•½ ë„êµ¬ ---
@tool
def process_and_summarize_webpage(url: str):
    """
    ì‚¬ìš©ìê°€ ì œê³µí•œ URLì˜ ì›¹í˜ì´ì§€ ì½˜í…ì¸ ë¥¼ ë¡œë“œí•˜ì—¬ ë¶„ì„í•˜ê³  ê·¸ ë‚´ìš©ì„ ìš”ì•½í•©ë‹ˆë‹¤.
    URLê³¼ ê´€ë ¨ëœ ì§ˆë¬¸ì— ë‹µí•˜ê±°ë‚˜ ë‚´ìš©ì„ ìš”ì•½í•´ì•¼ í•  ë•Œ ì‚¬ìš©í•´ì•¼ í•©ë‹ˆë‹¤.
    """
    try:
        loader = WebBaseLoader([url])
        docs = loader.load()
        summarize_chain = load_summarize_chain(llm=llm_for_tools, chain_type="stuff")
        summary = summarize_chain.invoke(docs)  # run ëŒ€ì‹  invoke ì‚¬ìš© (LangChain v0.2+)
        return summary
    except Exception as e:
        return f"ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}"


# --- ë„êµ¬ 4: ë²ˆì—­ ë„êµ¬ ---
@tool
def translate_text(text: str, target_language: str):
    """
    ì£¼ì–´ì§„ í…ìŠ¤íŠ¸ë¥¼ ì§€ì •ëœ ëŒ€ìƒ ì–¸ì–´ë¡œ ë²ˆì—­í•©ë‹ˆë‹¤.
    ì‚¬ìš©ìê°€ ë²ˆì—­ì„ ìš”ì²­í•  ë•Œ ì‚¬ìš©í•´ì•¼ í•©ë‹ˆë‹¤.
    """
    prompt = ChatPromptTemplate.from_template("Translate the following text to {target_language}:\n\n{text}")
    translation_chain = prompt | llm_for_tools | StrOutputParser()
    return translation_chain.invoke({"text": text, "target_language": target_language})
````

### 3. ì „ë¬¸í™”ëœ Agent ìƒì„±

````python
from langgraph.prebuilt import create_react_agent

#
# ê° Agentì— ëŒ€í•œ ì‹œìŠ¤í…œ í”„ë¡¬í”„íŠ¸ ì •ì˜
ml_system_prompt = """ë‹¹ì‹ ì€ ë¨¸ì‹ ëŸ¬ë‹ ì „ë¬¸ê°€ì…ë‹ˆë‹¤. 
ë¨¸ì‹ ëŸ¬ë‹, ë”¥ëŸ¬ë‹, AI ê´€ë ¨ ì§ˆë¬¸ì— ì „ë¬¸ì ìœ¼ë¡œ ë‹µë³€í•˜ì„¸ìš”.
ê´€ë ¨ ë„êµ¬ë¥¼ í™œìš©í•˜ì—¬ ì •í™•í•œ ì •ë³´ë¥¼ ì œê³µí•˜ì„¸ìš”."""

web_system_prompt = """ë‹¹ì‹ ì€ ì›¹ ê²€ìƒ‰ ì „ë¬¸ê°€ì…ë‹ˆë‹¤. 
ìµœì‹  ì •ë³´, ë‰´ìŠ¤, ì¼ë°˜ ìƒì‹ì— ëŒ€í•œ ì§ˆë¬¸ì— ë‹µë³€í•˜ì„¸ìš”.
ê²€ìƒ‰ ë„êµ¬ë¥¼ í™œìš©í•˜ì—¬ ìµœì‹  ì •ë³´ë¥¼ ì œê³µí•˜ì„¸ìš”."""

url_system_prompt = """ë‹¹ì‹ ì€ ì›¹í˜ì´ì§€ ë¶„ì„ ì „ë¬¸ê°€ì…ë‹ˆë‹¤. 
URLì´ í¬í•¨ëœ ì§ˆë¬¸ì— ëŒ€í•´ í•´ë‹¹ ì›¹í˜ì´ì§€ë¥¼ ë¶„ì„í•˜ê³  ìš”ì•½í•˜ì„¸ìš”.
ì›¹í˜ì´ì§€ ë¶„ì„ ë„êµ¬ë¥¼ í™œìš©í•˜ì„¸ìš”."""

translation_system_prompt = """ë‹¹ì‹ ì€ ë²ˆì—­ ì „ë¬¸ê°€ì…ë‹ˆë‹¤. 
ë‹¤ì–‘í•œ ì–¸ì–´ ê°„ ë²ˆì—­ì„ ì •í™•í•˜ê³  ìì—°ìŠ¤ëŸ½ê²Œ ìˆ˜í–‰í•˜ì„¸ìš”.
ë²ˆì—­ ë„êµ¬ë¥¼ í™œìš©í•˜ì—¬ ê³ í’ˆì§ˆ ë²ˆì—­ì„ ì œê³µí•˜ì„¸ìš”."""


# 1. ë¨¸ì‹ ëŸ¬ë‹ ì „ë¬¸ Agent
ml_prompt = ChatPromptTemplate.from_messages([("system", ml_system_prompt), ("placeholder", "{messages}")])
ml_agent = create_react_agent(llm_for_tools, [retriever_tool], prompt=ml_prompt)  # âœ… RAG ê²€ìƒ‰ ë„êµ¬

# 2. Web Search Agent - ì›¹ ê²€ìƒ‰ ì „ë¬¸
web_prompt = ChatPromptTemplate.from_messages([("system", web_system_prompt), ("placeholder", "{messages}")])
web_search_agent = create_react_agent(llm_for_tools, [tavily_tool], prompt=web_prompt)  # âœ… Tavily ê²€ìƒ‰ ë„êµ¬

# 3. URL Analysis Agent - ì›¹í˜ì´ì§€ ë¶„ì„ ì „ë¬¸
url_prompt = ChatPromptTemplate.from_messages([("system", url_system_prompt), ("placeholder", "{messages}")])
url_analysis_agent = create_react_agent(
    llm_for_tools, [process_and_summarize_webpage], prompt=url_prompt  # âœ… URL ë¶„ì„/ìš”ì•½ ë„êµ¬
)

# 4. Translation Agent - ë²ˆì—­ ì „ë¬¸
translation_prompt = ChatPromptTemplate.from_messages(
    [("system", translation_system_prompt), ("placeholder", "{messages}")]
)
translation_agent = create_react_agent(llm_for_tools, [translate_text], prompt=translation_prompt)  # âœ… ë²ˆì—­ ë„êµ¬
````

### 4. Supervisor
ì§ˆë¬¸ì˜ ìœ í˜•ì„ ë¶„ì„í•˜ê³ , ì˜¬ë°”ë¥¸ ì „ë¬¸ê°€ì—ê²Œ ì§ˆë¬¸ì„ í• ë‹¹

````python
# ============= Supervisor Agent =============
from langchain_core.output_parsers import JsonOutputParser
from pydantic import BaseModel, Field


class AgentSelection(BaseModel):
    next_agent: str = Field(description="ì„ íƒëœ ë‹¤ìŒ ì—ì´ì „íŠ¸ ì´ë¦„")
    reasoning: str = Field(description="ì„ íƒ ì´ìœ ")


def create_supervisor_agent():
    """ì§ˆë¬¸ì„ ë¶„ì„í•˜ê³  ì ì ˆí•œ ì „ë¬¸ Agentë¥¼ ì„ íƒí•˜ëŠ” Supervisor"""
    return ChatOpenAI(
        model="gpt-4o-mini",
        temperature=0,
        model_kwargs={"response_format": {"type": "json_object"}},  # âœ… JSON í˜•ì‹ ë³´ì¥
    )


supervisor_llm = create_supervisor_agent()
parser = JsonOutputParser(pydantic_object=AgentSelection)  # âœ… Supervisor ì¶œë ¥ íŒŒì„œ

# Supervisor ì²´ì¸ ìƒì„±
supervisor_chain = supervisor_llm | parser


def supervisor_node(state: AgentState) -> Dict[str, Any]:
    """Supervisorê°€ ì§ˆë¬¸ì„ ë¶„ì„í•˜ê³  ë‹¤ìŒ Agentë¥¼ ê²°ì •"""

    user_input = state.get("user_input", "")

    # JSON ì¶œë ¥ì„ ìœ„í•œ í”„ë¡¬í”„íŠ¸
    analysis_prompt = f"""ë‹¤ìŒ ì§ˆë¬¸ì„ ë¶„ì„í•˜ì—¬ ê°€ì¥ ì ì ˆí•œ Agentë¥¼ ì„ íƒí•˜ì„¸ìš”:

ì§ˆë¬¸: {user_input}

ì„ íƒ ê¸°ì¤€:
- ë¨¸ì‹ ëŸ¬ë‹ ë˜ëŠ” ë”¥ëŸ¬ë‹ì— ê´€ë ¨ëœ ì§ˆë¬¸ â†’ ml_agent
- ìµœì‹  ì •ë³´, ì¼ë°˜ ìƒì‹, í˜„ì¬ ë™í–¥ â†’ web_search_agent  
- URLì´ í¬í•¨ë¨ â†’ url_analysis_agent
- ë²ˆì—­ ìš”ì²­ â†’ translation_agent

ë°˜ë“œì‹œ JSON í˜•ì‹ìœ¼ë¡œ ì‘ë‹µí•˜ì„¸ìš”:
{{
    "agent": "ì„ íƒëœ_ì—ì´ì „íŠ¸_ì´ë¦„",
    "reasoning": "ì„ íƒ ì´ìœ  ì„¤ëª…"
}}"""

    try:
        parsed_output = supervisor_chain.invoke([HumanMessage(content=analysis_prompt)])
        next_agent = parsed_output["agent"]  # âœ… JSON ê²°ê³¼ì—ì„œ ì¶”ì¶œ
        reasoning = parsed_output["reasoning"]

    except Exception as e:
        print(f"Supervisor ì˜¤ë¥˜: {e}")
        next_agent = "web_search_agent"  # ì˜¤ë¥˜ ì‹œ ê¸°ë³¸ê°’
        reasoning = "íŒŒì‹± ì˜¤ë¥˜ë¡œ ì¸í•œ ê¸°ë³¸ ì„ íƒ"

    return {
        "next": next_agent,  # âœ… ë‹¤ìŒ ë…¸ë“œ
        "selection_reasoning": reasoning,  # âœ… ì„ íƒ ì´ìœ  ì €ì¥
    }
````

### 5. ê° ì—ì´ì „íŠ¸ ë…¸ë“œ

````python
from typing import Any, Dict

from langchain_core.messages import AIMessage, HumanMessage


def ml_agent_node(state: AgentState) -> Dict[str, Any]:
    """ML Agent ë…¸ë“œ - ë¨¸ì‹ ëŸ¬ë‹ ê´€ë ¨ ì§ˆë¬¸ ì²˜ë¦¬"""
    user_input = state.get("user_input", "")

    try:
        result = ml_agent.invoke({"messages": [HumanMessage(content=user_input)]})
        answer = result["messages"][-1].content if result.get("messages") else "ë‹µë³€ ìƒì„± ì‹¤íŒ¨"

        return {"rag_result": answer, "final_answer": answer, "messages": [AIMessage(content=answer)]}
    except Exception as e:
        error_msg = f"ML Agent ì²˜ë¦¬ ì˜¤ë¥˜: {str(e)}"
        return {"rag_result": error_msg, "final_answer": error_msg, "messages": [AIMessage(content=error_msg)]}


def web_search_agent_node(state: AgentState) -> Dict[str, Any]:
    """Web Search Agent ë…¸ë“œ - ì›¹ ê²€ìƒ‰ ì²˜ë¦¬"""
    user_input = state.get("user_input", "")

    try:
        result = web_search_agent.invoke({"messages": [HumanMessage(content=user_input)]})
        answer = result["messages"][-1].content if result.get("messages") else "ë‹µë³€ ìƒì„± ì‹¤íŒ¨"

        return {"web_result": answer, "final_answer": answer, "messages": [AIMessage(content=answer)]}
    except Exception as e:
        error_msg = f"Web Search Agent ì²˜ë¦¬ ì˜¤ë¥˜: {str(e)}"
        return {"web_result": error_msg, "final_answer": error_msg, "messages": [AIMessage(content=error_msg)]}


def url_analysis_agent_node(state: AgentState) -> Dict[str, Any]:
    """URL Analysis Agent ë…¸ë“œ - URL ë¶„ì„ ì²˜ë¦¬"""
    user_input = state.get("user_input", "")

    try:
        result = url_analysis_agent.invoke({"messages": [HumanMessage(content=user_input)]})
        answer = result["messages"][-1].content if result.get("messages") else "ë‹µë³€ ìƒì„± ì‹¤íŒ¨"

        return {"url_result": answer, "final_answer": answer, "messages": [AIMessage(content=answer)]}
    except Exception as e:
        error_msg = f"URL Analysis Agent ì²˜ë¦¬ ì˜¤ë¥˜: {str(e)}"
        return {"url_result": error_msg, "final_answer": error_msg, "messages": [AIMessage(content=error_msg)]}


def translation_agent_node(state: AgentState) -> Dict[str, Any]:
    """Translation Agent ë…¸ë“œ - ë²ˆì—­ ì²˜ë¦¬"""
    user_input = state.get("user_input", "")

    try:
        result = translation_agent.invoke({"messages": [HumanMessage(content=user_input)]})
        answer = result["messages"][-1].content if result.get("messages") else "ë‹µë³€ ìƒì„± ì‹¤íŒ¨"

        return {"translation_result": answer, "final_answer": answer, "messages": [AIMessage(content=answer)]}
    except Exception as e:
        error_msg = f"Translation Agent ì²˜ë¦¬ ì˜¤ë¥˜: {str(e)}"
        return {"translation_result": error_msg, "final_answer": error_msg, "messages": [AIMessage(content=error_msg)]}
````

### 6. Workflow êµ¬ì„±

````python
from langgraph.checkpoint.memory import MemorySaver
from langgraph.graph import END, START, StateGraph


def create_multi_agent_graph():
    """ë©€í‹° ì—ì´ì „íŠ¸ ì›Œí¬í”Œë¡œìš° ìƒì„±"""
    workflow = StateGraph(AgentState)

    # ë…¸ë“œ ì¶”ê°€
    workflow.add_node("supervisor", supervisor_node)
    workflow.add_node("ml_agent", ml_agent_node)
    workflow.add_node("web_search_agent", web_search_agent_node)
    workflow.add_node("url_analysis_agent", url_analysis_agent_node)
    workflow.add_node("translation_agent", translation_agent_node)

    # ë¼ìš°í„° í•¨ìˆ˜
    def supervisor_router(state: AgentState) -> str:
        """Supervisorì˜ ê²°ì •ì— ë”°ë¼ ë‹¤ìŒ ë…¸ë“œ ê²°ì •"""
        return state.get("next", "web_search_agent")  # ê¸°ë³¸ê°’ ì„¤ì •

    # ì—£ì§€ ì„¤ì •
    workflow.add_edge(START, "supervisor")

    # ì¡°ê±´ë¶€ ì—£ì§€ - supervisorì—ì„œ ê° agentë¡œ
    workflow.add_conditional_edges(
        source="supervisor",
        path=supervisor_router,  # âœ… supervisor_router í•¨ìˆ˜ ì‚¬ìš©
        path_map={  # âœ… Supervisor ê²°ê³¼ â†’ ì‹¤ì œ ë…¸ë“œ ë§¤í•‘
            "ml_agent": "ml_agent",
            "web_search_agent": "web_search_agent",
            "url_analysis_agent": "url_analysis_agent",
            "translation_agent": "translation_agent",
        },
    )

    # ê° agentì—ì„œ ENDë¡œ
    workflow.add_edge("ml_agent", END)
    workflow.add_edge("web_search_agent", END)
    workflow.add_edge("url_analysis_agent", END)
    workflow.add_edge("translation_agent", END)

    # ë©”ëª¨ë¦¬ ì²´í¬í¬ì¸í„° ì¶”ê°€
    memory = MemorySaver()
    app = workflow.compile(checkpointer=memory)
    return app
````

````python
# ì›Œí¬í”Œë¡œìš° ìƒì„±
multi_agent_system = create_multi_agent_graph()
````

````python
import uuid


# ì‹¤í–‰ í•¨ìˆ˜
def run_multi_agent_system(question: str, config: dict = None):
    """ë©€í‹° ì—ì´ì „íŠ¸ ì‹œìŠ¤í…œ ì‹¤í–‰"""

    # ê° ì‹¤í–‰ë§ˆë‹¤ ê³ ìœ í•œ thread_id ìƒì„±
    if config is None:
        config = {"configurable": {"thread_id": str(uuid.uuid4())}}

    # ì´ˆê¸° ìƒíƒœ ì„¤ì •
    initial_state = {
        "messages": [],
        "next": "",
        "user_input": question,
        "selection_reasoning": "",
        "ml_result": "",
        "web_result": "",
        "url_result": "",
        "translation_result": "",
        "final_answer": "",
    }

    print(f"\n{'='*50}")
    print(f"ì§ˆë¬¸: {question}")
    print(f"{'='*50}")

    try:
        # ì›Œí¬í”Œë¡œìš° ì‹¤í–‰
        result = multi_agent_system.invoke(initial_state, config)

        # final_answerê°€ ì„¤ì •ë˜ì–´ ìˆìœ¼ë©´ ìš°ì„  ì‚¬ìš©
        final_answer = result.get("final_answer") or (
            result.get("rag_result")
            or result.get("web_result")
            or result.get("url_result")
            or result.get("translation_result")
            or "ë‹µë³€ì„ ìƒì„±í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
        )

        print(f"\në‹µë³€: {final_answer}")
        print(f"{'='*50}\n")

        return {
            **result,
            "final_answer": final_answer,
            "next": result.get("next"),
            "selection_reasoning": result.get("selection_reasoning"),
        }

    except Exception as e:
        error_msg = f"ì‹œìŠ¤í…œ ì˜¤ë¥˜: {str(e)}"
        print(f"\nì˜¤ë¥˜: {error_msg}")
        print(f"{'='*50}\n")

        return {"error": error_msg, "final_answer": error_msg}
````

### FYI. workflow ë¥¼ ì‹œê°ì ìœ¼ë¡œ í™•ì¸ (Timeout ì—ëŸ¬ê°€ ë°œìƒí•  ê²½ìš°)

âœ¨ ë³µì¡í•œ `LangGraph` ê·¸ë˜í”„ë¥¼ ì£¼í”¼í„° ë…¸íŠ¸ë¶ì— í‘œì‹œí•  ë•Œ, **`mermaid.ink` APIì˜ ì‹œê°„ ì´ˆê³¼** ì˜¤ë¥˜ë¥¼ í”¼í•˜ê¸° ìœ„í•´ **ë¡œì»¬ ë Œë”ë§ ë°©ì‹**ì„ ì‚¬ìš©í–ˆìŠµë‹ˆë‹¤.

* **í•´ê²° ë°©ë²•**: `compile()` ì´í›„ `get_graph().draw_mermaid_png()` ë©”ì„œë“œì— **`draw_method=MermaidDrawMethod.PYPPETEER`** ì¸ìë¥¼ ì „ë‹¬í•˜ì—¬ ë¡œì»¬ì—ì„œ ë Œë”ë§í–ˆìŠµë‹ˆë‹¤.

````python
import asyncio
import os
from typing import Union

from langchain_core.runnables.graph_mermaid import MermaidDrawMethod
from langgraph.graph import StateGraph


def save_workflow_graph(workflow: Union[StateGraph, object], filename: str, async_mode: bool = False) -> None:
    """
    Mermaid ê·¸ë˜í”„ PNGë¥¼ íŒŒì¼ë¡œ ì €ì¥í•˜ëŠ” í•¨ìˆ˜.

    Args:
        workflow: StateGraph ë˜ëŠ” multi-agent workflow ê°ì²´
        filename: ì €ì¥í•  íŒŒì¼ëª… (ìë™ìœ¼ë¡œ .png í™•ì¥ì ì¶”ê°€)
        async_mode: Trueë©´ await í•„ìš” ì‹œ ë¹„ë™ê¸° ë°©ì‹ìœ¼ë¡œ ì‹¤í–‰
    """
    if not filename.lower().endswith(".png"):
        filename += ".png"

    try:
        if async_mode:
            import nest_asyncio

            nest_asyncio.apply()  # Notebook ë‚´ë¶€ ì´ë²¤íŠ¸ ë£¨í”„ íŒ¨ì¹˜

            async def _draw_async():
                return await workflow.get_graph().draw_mermaid_png(draw_method=MermaidDrawMethod.PYPPETEER)

            loop = asyncio.get_event_loop()
            graph_png = loop.run_until_complete(_draw_async())
        else:
            # ë™ê¸° í˜¸ì¶œ
            graph_png = workflow.get_graph().draw_mermaid_png(draw_method=MermaidDrawMethod.PYPPETEER)

        with open(filename, "wb") as f:
            f.write(graph_png)

        print(f"ê·¸ë˜í”„ê°€ '{filename}' íŒŒì¼ë¡œ ì„±ê³µì ìœ¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

    except Exception as e:
        print(f"ê·¸ë˜í”„ ì €ì¥ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
````
